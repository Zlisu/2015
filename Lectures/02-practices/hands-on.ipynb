```python
def f2(gr):
    #print(gr.shape)
    #print(type(gr))
    #print(gr.iloc[:, 2])
    
    arr = gr.iloc[:, 2]
    if (arr[arr == 'M'].count()) > (arr[arr == 'F'].count()):
        return pd.Series('Male Dominated')
    elif (arr[arr == 'M'].count()) < (arr[arr == 'F'].count()):
        return pd.Series('femal Dominated')
    else:
        return pd.Series('equally distributed between genders')
```
