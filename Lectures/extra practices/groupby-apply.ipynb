```python
import seaborn as sns
import pandas as pd

iris_data = sns.load_dataset('iris')
df = iris_data.head(5).copy()
df = pd.concat([df, iris_data.iloc[50:55]])
df = pd.concat([df, iris_data.iloc[100:105]])
```

```python
df.groupby('species').sum()
```

```python
multicol_sum = df.groupby(['species', 'petal_width']).sum()
multicol_sum
```

## applying functions

```python
df.groupby('species').apply(lambda gr:gr.sum())
```

```python
def my_sum(gr):
    return gr.sum()

df.groupby('species').apply(my_sum)
```

```python
def foo(gr):
    print(type(gr))
    return None

df.groupby('species').apply(foo)
```

```python
df.groupby('species')['petal_length'].apply(foo)
```

```python
def foo(gr):
    print(gr, '\n')

df.groupby('species').apply(func=foo)
```

```python
def foo(gr):
    return pd.Series("This is a test")

df.groupby('species').apply(func=foo)
```

```python
myseries = pd.Series(
    data = ['one', 'two', '3'],
    index = ['a', 'b', 'c']
)

myseries
```

```python
def foo(gr):
    return myseries

# a series is returned by the applied function
df.groupby(['species', 'petal_width']).apply(func=foo)
```

```python
def foo(gr):
    return pd.DataFrame(myseries)

df.groupby(['species', 'petal_width']).apply(func=foo)
```
